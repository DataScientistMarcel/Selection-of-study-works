### Schritt 2: Neuen Dataframe mit ausgewählten Spalten erstellen und Zeitwerte in Datumsobjekte 'datetime64' konvertieren

```python
df_data_combined1 = df_data_combined[['Date Published', 'Publication Date', 'material-class-label']].copy()
```

```python
# Zusammenführen der beiden Spalten 'Date Published' und 'Publication Date' in eine Gesamtspalte
df_data_combined1['Combined Date'] = df_data_combined1['Date Published'].combine_first(df_data_combined1['Publication Date'])
```

```python
# Anzahl an ungültigen Werten anzeigen lassen
invalid_dates = df_data_combined1['Combined Date'].isna().sum()
print(f"Ungültige oder fehlende Datumswerte: {invalid_dates}")
```

```python
df_data_combined1['Combined Date'] = pd.to_datetime(df_data_combined1['Combined Date'], errors='coerce')
# errors='coerce' --> Umwandlung von ungültigen Werten in NaT (Not a Time)
```

### Schritt 3: Visualisierung als Timeline

```python
df_data_combined1['material-class-label']
```

```python
df_data_combined1['Year'] = df_data_combined1['Combined Date'].dt.year
```

```python
# Neuen Dataframe aus Subset erstellen, welches nur das Jahr und das Klassenlabel beinhaltet
df_filtered = df_data_combined1.dropna(subset=['Year', 'material-class-label'])
```

```python
# Zeitreihe erstellen
timeline = df_filtered.groupby(['Year', 'material-class-label']).size().reset_index(name='Count')
```

```python
# Farben für die Rohstoffe definieren
farben = {
    'lithium': '#1f77b4',   # Blau
    'sodium': '#ff7f0e',   # Orange
    'zinc': '#2ca02c',      # Grün
    'calcium': '#d62728',    # Rot
    'magnesium': '#ffff00',   # Gelb
    'aluminum': '#8c564b'     # Braun
}
```

```python
# Visualisierung erstellen
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
sns.lineplot(data=timeline, x='Year', y='Count', hue='material-class-label', palette=farben, 
             linewidth=2.5, marker='o', markersize=8)

plt.title('Veröffentlichungen pro Jahr nach Rohstoff', fontsize=18)
plt.xlabel('Jahr', fontsize=14)
plt.ylabel('Anzahl Veröffentlichungen', fontsize=14)
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12) 
plt.grid(True)
plt.legend(title='Rohstoff', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12, title_fontsize=13)
plt.tight_layout()
# plt.savefig('timeline_rohstoffe.png', dpi=300)
plt.show()
```
