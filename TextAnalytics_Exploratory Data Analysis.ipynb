### 1.5) Verteilung der Daten darstellen (Balkendiagramm, Boxplot, Histogramm)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Zähle die Anzahl der Artikel pro Land
source_country_counts = df_scholarly_articles1['Source Country'].value_counts()
print(source_country_counts.to_string())
```

```python
sns.set(font_scale=1.1)  # Vergrößert die Schriftgröße der Achsenticks

# Horizontales Balkendiagramm
plt.figure(figsize=(10, 12))
sns.barplot(y=source_country_counts.index, x=source_country_counts.values, color='steelblue')
plt.title('Verteilung der wissenschaftl. Artikel nach Herkunftsland')
plt.xlabel('Anzahl der Artikel')
plt.ylabel('Herkunftsland')
plt.tight_layout()
plt.show()
```

```python
# Boxplot mit Ausreißern
df_scholarly_articles1['Abstract_Length'].plot(kind='box', vert=False)
plt.show()
```

```python
# Boxplot ohne Ausreißer zur besseren Darstellung der Box
df_scholarly_articles1.boxplot(column='Abstract_Length', vert=False, showfliers=False)
plt.show()
```

```python
# Histogramm
df_scholarly_articles1['Abstract_Length'].plot(kind='hist', bins=40)

# Achsenbeschriftungen hinzufügen
plt.xlabel('Anzahl Zeichen')
plt.ylabel('Häufigkeit')

plt.show()
```

### 1.6) Visualisierung von wichtigen Keywords mittels Word Cloud

```python
from collections import Counter
```

```python
# Transformierung der Abstracts in eine Liste von Wörtern (Leerzeichen ' ' als Splitpunkt)
# Sicherstellung derselben zufälligen 1000 Abstracts durch festen Zufalls-Seed
word_list = df_scholarly_articles1['Abstract'].sample(n=1000, random_state=42).dropna().str.split(' ').sum()
```

```python
# Zählung der Wörter
word_counts = Counter(word_list)
word_counts.most_common(20)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Wortliste in einen String umwandeln
text = ' '.join(word_list)

# Wordcloud erzeugen
wordcloud = WordCloud(stopwords=STOPWORDS, background_color='white', width=800, height=400).generate(text)
```

```python
# Wordcloud anzeigen
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Achsen werden nicht benötigt
# plt.savefig('scholarly_articles_abstracts_without_custom_stopwords.png', bbox_inches='tight', dpi=200) # Bild speichern
plt.show()
```

```python
# Erstellung einer neuen Wortliste aus denselben zufälligen 1000 Abstracts wie zuvor
word_list_without_stopwords = df_scholarly_articles1['Abstract'].sample(n=1000, random_state=42).dropna().str.split(' ').sum()

# Eigene Stoppwörter definieren und zur Liste mit den herkömmlichen Stoppwörtern hinzufügen
custom_stopwords = set(STOPWORDS).union({
    'The', 'jats', 'sub', 'sup', 'et al', 'et', 'al', 'p', 'italic', 'due', 'used', 'using', 'however', 'however,', 
    'batteries.', 'study', 'results', 'data', 'research', 'paper', 'method', 'methods', 'conclusion'
})

# Stoppwörter entfernen
words_without_stopwords = [word for word in word_list if word.lower() not in custom_stopwords]

Counter(words_without_stopwords).most_common(20)
```

```python
# Wortliste in einen String umwandeln
text_without_stopwords = ' '.join(words_without_stopwords)

wordcloud = WordCloud(stopwords=custom_stopwords, background_color='white', width=800, height=400).generate(text_without_stopwords)

# Plot wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Achsen werden nicht benötigt
# plt.savefig('scholarly_articles_abstracts_with_custom_stopwords.png', bbox_inches='tight', dpi=200)
plt.show()
```
